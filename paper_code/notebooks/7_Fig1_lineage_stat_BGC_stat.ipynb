```python
# 统计每个phylum-genus下的BGC的平均总数
mean_sum_result = new_merged_df.groupby(by=["phylum", "genus"])["Sum"].mean()
mean_sum_result
```

```python
import numpy as np

# 统计每组的行数
group_sizes = new_merged_df.groupby(by=["phylum", "genus"]).size()
group_sizes.name = "size"
group_sizes

# 对 group_sizes 取 log10
group_sizes_log10 = np.log10(group_sizes)

group_sizes_log10
```

```python
counts = new_merged_df['16S_ID'].value_counts()
max_count = counts.max()
most_frequent_count = counts[counts == max_count]
most_frequent_count.index[0]
```

```python
# 计算每个分组中的 "16S_ID" 出现次数
def most_frequent_16S(group):
    # 计算每个 "16S_ID" 的出现次数
    counts = group['16S_ID'].value_counts()
    
    # 获取最大出现次数
    max_count = counts.max()
    
    # 获取所有出现次数为最大值的 16S_ID
    most_frequent_count = counts[counts == max_count]
    
    # 如果只有一个 16S_ID 出现次数最多，返回该 16S_ID；如果有多个，则随机选择一个
    most_frequent = most_frequent_count.sample(n=1, random_state=2).index[0]  # 随机选择一个
    
    return most_frequent

# 统计每个phylum-genus下的代表性16S
most_frequent_16S_result = new_merged_df.groupby(by=["phylum", "genus"]).apply(most_frequent_16S)
most_frequent_16S_result
```

```python
# 查找重复的值
duplicates = most_frequent_16S_result[most_frequent_16S_result.duplicated(keep=False)]

# 打印重复的值
print(duplicates)
len(duplicates)
```

```python
# 3. 合并两个结果
merged_result = pd.concat([most_frequent_16S_result, mean_sum_result, group_sizes_log10], axis=1)
merged_result
```

```python
merged_result.to_csv("data/Fig1C.csv")
```

```python
target_seq = most_frequent_16S_result.to_list()
target_seq
```

```python
from Bio import SeqIO
from pathlib import Path

# 读取fasta文件
def read_fasta(fasta_path):
    records = list(SeqIO.parse(fasta_path, 'fasta'))
    return records

# 找出代表性序列用于建树
def get_target_records(records):
    keep_records = []
    for record in records:
        if record.id in target_seq:
            keep_records.append(record)
    return keep_records

records = read_fasta("data/filtered_intergrated_16SrRNA.fasta")
target_records = get_target_records(records)

# 保存序列
output_file = Path('data/Fig1C_genus_rep_16SrRNA.fasta')
SeqIO.write(target_records, output_file, 'fasta')
```
